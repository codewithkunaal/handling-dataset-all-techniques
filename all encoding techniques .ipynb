## one hot encoding with many catagories

if there are many catagories replace them with only frequent catagories 
eg. if there are 50 catagories use only top (9)+ 1 catagories for encoding and use one last catagory for rest of catagories

```python
df = pd.read_csv("mercedes.csv",usecols=["X0","X1","X2","X3","X4"])
```

```python
df.head()
```

```python
df["X0"].value_counts()
```

```python
for i in df.columns :
    print(f" total catagories in {i} are {len(df[i].unique())}")
```

```python
catagories = df["X0"].value_counts(ascending=False).head(15).index
```

```python
catagories
```

```python
import numpy as np 
```

```python
for i in catagories:
    df[i]=np.where(df["X0"]==i,1,0)
    
```

```python
df.head()
# df.columns
```

```python
df.drop(["X1","X2","X3","X4"],axis=1)
# catagories.index = pd.Index(df.index.values-1) 
```

```python
catagories
```

## mean encoding - nominal

taking mean of each catagory and replacing it with mean

```python
df = pd.read_csv("titanic.csv",usecols=["Survived","Cabin"])
```

```python
df.columns
```

```python
df.head()
```

```python
df["Cabin"].fillna("M",inplace=True)
```

```python
df["Cabin"] = df["Cabin"].astype(str).str[0]
```

```python
df
```

```python
df.Cabin.unique()
```

```python
labels=df.groupby(["Cabin"])["Survived"].mean().sort_values(ascending=True)
```

```python
df["Cabin_mean"]=df["Cabin"].map(labels)
```

```python
df.drop(["Cabin"],axis=1)
```

## target guided encoding 


encode according to output label 


```python
df = pd.read_csv("titanic.csv",usecols=["Cabin","Survived"])
```

```python
df=df.fillna("M")
```

```python
df["cabin_en"]=df["Cabin"].astype(str).str[0]
```

```python
df
```

```python
df.cabin_en.unique()
```

```python
df.groupby(["cabin_en"])["Survived"].mean()
```

```python
labels = df.groupby(['cabin_en'])['Survived'].mean().sort_values().index
#sorting for importance of each catagory
```

```python
labels
```

```python
encode = {j:i for i,j in enumerate(labels,0)}
```

```python
encode
```

```python
df["encoded_cabin"]=df["cabin_en"].map(encode)
```

```python
df
```

# probability ratio encoding

depending on output of each catagory we encode the colm with probability of the catagorical value
works well for two catagorical labels                           
formula = class2 = 1 - class1                         

        probability = class1/class2

```python
import pandas as pd
df = pd.read_csv("titanic.csv",usecols=["Survived","Cabin"])
```

```python
df.head()
```

```python
df["Cabin"]=df["Cabin"].fillna("M")
```

```python
df
```

```python
df["Cabin_en"]=df["Cabin"].astype(str).str[0]
```

```python
df
```

```python
proba =df.groupby(["Cabin_en"])["Survived"].mean()
```

```python

proba=pd.DataFrame(proba)
```

```python
proba
```

```python
df
```

```python
proba["Survived"]
```

```python
proba["not_survived"]=1-proba["Survived"]
```

```python
proba
```

```python
proba["ratio"] = proba["Survived"]/proba["not_survived"]
```

```python
proba
```

```python
dict=proba["ratio"].to_dict()
```

```python
dict
```

```python
df["Cabin_encoded"] = df["Cabin_en"].map(dict)
```
